```python
# Titanic Data Exploration and Feature Engineering
This notebook explores the Titanic dataset and prepares features for predictive modeling.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load data
df = pd.read_csv('../data/train.csv')

# Missing Value Handling
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Create FamilySize
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Extract Titles
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

# Visualize survival by Title (Required by Part 2.5 of assignment)
sns.barplot(x='Title', y='Survived', data=df)
plt.title('Survival Rate by Title')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Quick prep for importance check
X = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'FamilySize']]
y = df['Survived']

model = RandomForestClassifier()
model.fit(X, y)

features = pd.Series(model.feature_importances_, index=X.columns)
features.sort_values().plot(kind='barh')
plt.title('Feature Importance')
plt.show()
```
